### Matrix Multiplication
1. Element-wise
2. Matrix multiplication

Rules:
1. Inner dimensions must match:
* `(3, 2) @ (2, 3)` will work
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
2. Resulting matrix has the shape of the outer dimensions:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
torch.matmul(torch.rand(2, 3), torch.rand(3, 5))
```

```python
# element wise
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
%%time
torch.matmul(tensor, tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

### One of the most common errors in deep learning: shape errors

```python
# shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
tensor_A.shape, tensor_B.shape
```

```python
# torch.mm(tensor_A, tensor_B) # torch.mm is the same as torch.matmul (its an alias for writing less code)

# To fix tensor shape, we can manipulate shape of our tensors using a `transpose`
tensor_B.T, tensor_B.T.shape
```

```python
# matrix multiplication works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}, tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- Inner dimensions must match")
print("Output:")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## Reshape, stack, squeeze, unsqueeze tensors

* Reshape - reshapes an input tensor to a defined shape
* View - return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stack - Combine multiple tensors on top of each other (vstack) or side-by-side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Create a tensor
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because view of a tensor shares the same memory as the original tensor)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
# x_stacked = torch.stack([x, x, x, x], dim=0) # vertical
x_stacked = torch.stack([x, x, x, x], dim=1) # horizontal
x_stacked, x_stacked.shape
```

```python
x_reshaped, x_reshaped.shape
```

```python
x_squeezed = x_reshaped.squeeze()
x_squeezed, x_squeezed.shape
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed, x_unsqueezed.shape
```

```python
# torch.permute rearranges the dimensions of the target tensor in a specified order --> returns a view
x_original = torch.rand(224, 224, 3) # H, W, C
# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = torch.permute(x_original, dims=(2, 0, 1)) # C, H, W
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
# x_permuted will share the same memory as x_original as its just a view
x_original[0, 0, 0] = 72138
print(f"Before: x_original = {x_original[0, 0, 0]}, x_permuted = {x_permuted[0, 0, 0]}")
x_permuted[0, 0, 0] = 22
print(f"After: x_original = {x_original[0, 0, 0]}, x_permuted = {x_permuted[0, 0, 0]}")
```

## Indexing (selecting data from tensors)

Indexing with PyTorch is similar to indexing with NumPy

```python
# Create a tensor
x = torch.arange(1, 10).reshape(1, 3, 3)
x, x.shape
```

```python
# Lets index on our tensor
x[0]
```

```python
# Lets index on the middle brack (dim = 1)
x[0][0] # or x[0, 0]
```

```python
# Lets index on the most inner bracket (dim = 2)
x[0, 2, 2]
```

```python
# You can also use ":" to select "all" of the target dimension
x[:,0]
```

```python
# Get all values of 0th and 1st dimensions but only index 1 of the 2nd dimension
x[:, :, 1]
```

```python
# Get all values of 0th dim but only the 1 index of the 1st and 2nd dim
x[:, 1, 1]
```

```python
# Get index 0 of 0th and 1st dim and all values of 2nd dim
x[0, 0, :]
```

```python
# Index on x to return 3, 6, 9
x[0, :, 2]
```

## PyTorch tensors & NumPy

* NumPy -> PyTorch: `torch.from_numpy(ndarray)
* PyTorch -> NumPy: `torch.Tensor.numpy()`

Default numpy datatype: float64
Default tensor datatype: float32

```python
# NumPy array to tensor
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when converting from numpy -> pytorch, Pytorch reflects numpy's default datatype float64 unless specified otherwise
array, tensor
```

```python
# Change the value of array, does not change value of `tensor`
array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

## Reproducibility (trying to take random out of random)

To reduce the randomness in NN and PyTorch, comes the concept of random seed. Essentially, what the random seed does is flavor the randomness.

```python
# Create 2 random tensors

random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Lets make random but reproducible tensors
# Set random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED) # use this everything when you need to create randomness
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
